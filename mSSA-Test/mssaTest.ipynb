```python
import pandas as pd
import numpy as np
from mssa.mssa import mSSA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mixturets_var.csv")
model = mSSA()
model.update_model(df.loc[:,['ts']]) 
df.plot(x="time", y="ts")
```

```python
prediction = model.predict('ts',1000)
```

```python
prediction = model.predict('ts', 100001,100100)
```
